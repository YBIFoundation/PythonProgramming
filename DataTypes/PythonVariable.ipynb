# **Variable Naming**

- The first character of the variable must be an alphabet or underscore ( _ ).
- All the characters except the first character may be an alphabet of lower-case(a-z), upper-case (A-Z), underscore, or digit (0-9)
- Identifier name must not contain any white-space, or special character (!, @, #, %, ^, &, *).
- Identifier name must not be similar to any keyword defined in the language.

```python
# allowed naming
var = 1
```

```python
# allowed naming
_var = 1
```

```python
# allowed naming
_2var = 1
```

```python
# not allowed naming
*var = 1
```

```python
# not allowed naming
2var = 1
```

```python
# not allowed due to reserved keywords
class = 1
```

```python
# reserved key words
import keyword
keyword.kwlist
```

# **Variables**
- A variable is created the moment you first assign a value to it.

- Variables do not need to be declared with any particular type, and can even change type after they have been set.

```python
# integer
int = 2
type(int)
```

```python
# float
flt = 3.2
type(flt)
```

```python
# complex
cpx = 3+2j
type(cpx)
```

```python
# bool
bl = True
type(bl)
```

```python
# list
lt = [1,2.2,'ybi', True]
type(lt)
```

```python
# tupple
tp = (1,2.2,'ybi', True)
type(tp)
```

```python
# string
stg = 'ybi'
type(stg)
```

```python
# set
st = {1,2.2,'ybi', True, False}
type(st)
```

```python
# frozen set is similar to sets except that frozensets are immutable
fst = frozenset(st)
type(fst)
```

```python
# dictionary
dt = {'name': 'ybi', 1: 2, 'status':True}
type(dt)
```
